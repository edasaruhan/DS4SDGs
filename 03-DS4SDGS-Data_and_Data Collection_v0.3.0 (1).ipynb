## 1.2 General Characteristics of Data <a class="anchor" id="one-one"></a>

Data or Data sets possess a multitude of characteristics are important when working with specific types of data. Each of which has its own implication for data preprocessing, analysis, and modeling.

Here we will discuss 3 general characteristics of data sets: **Dimensionality, Sparsity,** and **Resolution**.



**What is Dimensionality?**

* The dimensionality of a data set is the number of attributes that the samples of the dataset have.
* High dimensionality can be considered as rich information of the dataset when it is compared to low dimensionality.
* High dimensionality may come cause complexity; however it can be solved with various techniques to enhance the accuracy of the model.



**What is Sparsity?**

* When most of the attributes of each sample in the data set has the value 0 (not N\A or NaN value), it is called as sparse data
* Sparsity of the data set is the gap caused by these samples
* Sparse data cannot be counted as missing value since it has the value 0
* It may cause failure while training machine learning models and effects its accuracy

**What is Resolution?**

* Resolution of the data set means that how finely data was collected
* If the collected data can represent the pattern, it has a high resolution. To give an example, if atmospheric pressure data was collected on a scale of hours it would provide more patterns while collecting it on monthly scale cause missing patterns.

## 1.3 Types of Datasets <a class="anchor" id="one-two"></a>

Data can be classified based on various parameters. In the context of data models and organization, data can be broadly categorized into three types: **Record Data, Graph-based Data,** and **Ordered Data**. These classifications help in representing the relationships, structures, and the nature of data.

### 1.3.1 **Record Data**

Record data consists of a collection of records, where each record is a collection of fields. Each field represents an attribute, and each record represents an entity or an object. The fields can be of various types, such as numerical, textual, etc.
<br>
- The majority of data mining work assumes that data is a collection of records, also known as data objects.

- The most basic form of record data has no explicit relationship among records or data fields, and every record has the same set of attributes.
- Record data is usually stored either in flat files or in relational databases (as tables with rows and columns, where each row represents a unique record, and each column represents an attribute of that record).

- Some of the variations of record data include **transaction or market basket data, the data matrix**, and the **sparse data matrix**.

- Record data is widely used in traditional relational databases and is suitable for structured data where relationships between different entities are defined using keys.

### 1.3.2 **Graph-based Data**

Graph-based data is represented using nodes and edges, where nodes represent entities (data objects), and edges represent the relationships between them. Graphs can be directed or undirected, weighted or unweighted.

* Graph-based data is used to represent complex relationships, especially when the relationships have equal or more importance as the entities themselves. They're common in network analysis, social network analysis, routing, etc.

* Usage Example of Graph-based data: Social networks where individuals are represented as nodes and friendships as edges, or web pages where nodes represent pages, and edges represent hyperlinks.

### 1.3.3 **Ordered Data**

Ordered data has an inherent sequence in its structure. The order or sequence in the data is vital, and changing this order can alter the meaning or interpretation of the data.

- **Sequential Data:** Sequential data refers to a collection where each record has a time association. This type of data has an inherent order.

- **Time Series Data:** Time series data is a special type of sequential data in which each record represents a series of measurements taken over time. It emphasizes the chronological ordering of data points.

- **Sequence Data:** Sequence data consists of a data set that is an ordered arrangement of individual entities, such as a sequence of words or letters. While similar to sequential data, sequence data may or may not have a time component, and the ordering itself is the significant aspect.

- **Spatial Data:** Spatial data pertains to objects that have spatial attributes, such as positions or areas, in addition to other types of attributes. This data type captures the geographical or spatial relationships between objects.

## **1.6 Data Formats and File Types**

Let's now explore various common data formats that are essential in the field of data science. Understanding these formats is crucial for importing, exporting, and manipulating data across different platforms and tools.

**CSV (Comma Seperated Values) files:**

- Comma separated values. It has a plaint text file that contains list of data.
- Each line represents a row, and commas separate the values (columns).
- CSV also known as flat file, is widely used due to its simplicity and wide support.

**Excel (XLSX) files:**
- XLS files are Binary Interchange File Format used to store spreadsheet data and are proprietary to Microsoft.
- Supports text, numbers, formulas, and a wide array of formatting options.
- Commonly used in business and academia for data storage and manipulation.

**JSON (JavaScript Object Notation)**
- A lightweight data interchange format.
- Uses human-readable text to represent structured data.
- Often used in web applications to transmit data between server and client.

**Pickled files:**
- Pickled files contain serialized data.

**HDF5 files:**
- Hierarchical Data Formats (HDF) used to store large amount of data efficiently. Highly used in data analysis as it supports a multitude of datatypes with great flexibility.

````python
# Establishing a connection to the SQLite database
conn = sqlite3.connect('ch3_sqlite.db')

# Executing an SQL query to fetch all records from the SDG table
cursor = conn.execute("SELECT ID,SDG_NUMBER,Goal from SDG")

# Iterating over the fetched records and printing each one of them
for row in cursor:
   print("ID = ", row[0])
   print("SDG_NUMBER = ", row[1])
   print("Goal = ", row[2], "\n")

# Printing a message indicating successful operation completion
print("Operation done successfully")

# Closing the connection to the database
conn.close()
````

## 3.2 Web Scraping  <a class="anchor" id="three-one"></a>


Web scraping is the process of extracting data from a website, a technique that taps into the vast resources available on the internet, including text, images, voice, video, and more. This data may be needed for further computation, and web scraping provides an efficient means to obtain it.

Web scraping is a powerful tool for data collection, but it comes with both technical challenges and ethical responsibilities. Understanding the tools and methodologies is essential.

The advantages of web scraping include:

* Cost-effectiveness, as the materials are sourced from the internet
* Low maintenance and speed in automated data collection
* Ease of implementation, with the right technical understanding

However, web scraping also has some disadvantages, such as:

* Time consumption in establishing baseline code
* Difficulty in analysis without proper technical knowledge
* Potential barriers due to protection policies on certain websites


Webscraping typically involves making HTTP requests to a website to retrieve the HTML content.

- It also supports sending extra information to a web server through parameters and headers, encoding the server responses, detecting errors, and handling redirects.

### 3.2.1 Importing Flat files from the Web

We can download and read flat files (such as CSV, TSV, or plain text files) directly from a web URL. This can be a way to access publicly available data hosted online without needing to manually download the file to your local system.

In the code example below, this process is carried out using the `requests library` to send an HTTP GET request to the specified URL and pandas to read the CSV data into a

````python
import pandas as pd
import io
import requests

# URL of the CSV file
url = (
    r'https://raw.githubusercontent.com/SDG-AI-Lab/Digital_Technologies_Radar/'
    'develop/src/assets/csv/technology_radar_dataset_updated_v8.csv'
)

try:
    r = requests.get(url)
    r.raise_for_status()  # Raise exception for failure

    # Read content into DataFrame
    df = pd.read_csv(io.StringIO(r.text))

except requests.RequestException as e:
    print(f"Error fetching data: {e}")

# Display first 3 rows
df.head(3)
````

### 3.2.2 Importing Non-flat files from the Web

* **2. Understanding the Webpage Structure:**
By inspecting elements, you gain insight into the hierarchy and layout of the webpage's HTML. This understanding is essential for writing accurate and efficient web scraping code.

* **3. Debugging and Troubleshooting:**
If your web scraping code isn't working as expected, inspecting elements can help you troubleshoot the issue. You can verify whether the elements you're targeting exist on the page, have the expected tags, classes, or IDs, and contain the data you're looking for.

* **4. Avoiding Unwanted Data:**
Inspecting elements allows you to precisely target the data you want to scrape and avoid extracting unnecessary or irrelevant information.

## 3.3 Performing requests with URL Library (urllib)

 The `urllib` library is a module in Python's standard library that provides functions and classes to work with URLs. It is used to fetch data across the web, handle URL components, encode and decode query strings, and more.

 In the example below, we show how to perform an HTTP request using Python's urllib library.



````python
# Importing necessary functions from the urllib.request module
from urllib.request import urlopen, Request

# Specifying the URL that you want to request
url = "https://sdgs.un.org/goals"
# Creating an HTTP request object for the specified URL
request = Request(url)

# Sending the request to the URL and receiving the response
response = urlopen(request)
# Reading the HTML content from the response
html = response.read()

# Printing the type of the response object to confirm the request was successful
print(type(response))

# Printing the HTML content
print(html.decode('utf-8'))  # Decode the bytes to string for better readability

# Closing the response to free up resources
response.close()
````

**Import libraries**

````python
# Import the 'requests' library for making HTTP requests
import requests

# Import the 'BeautifulSoup' class from the 'bs4' library for parsing HTML
from bs4 import BeautifulSoup

# Import the 'pandas' library for data manipulation and analysis
import pandas as pd
````

**Query information**

````python
months = ['01','02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
year = '2015'
city = 'Ankara/Central'
country = "Turkey"
````

**Table parameters inside the website table**

````python
parameters = ['Temperature avarage, Fahrenheit / Celsius', 'Temperature maximum, Fahrenheit / Celsius', 'Temperature minimum, Fahrenheit / Celsius', 'Precipitation in / cm', 'Average daily wind speed (miles/hour)',
              'Visibility in miles', 'Dew point, Fahrenheit / Celsius', 'Maximum sustained wind speed (miles/hour)', 'Maximum wind gust reported for the day', 'Sea level pressure in mBARs',
              'Station pressure in mBARs', 'Snow depth in inches', 'Weather type']
````

````python
records = {'City': [], 'Year': [], 'Month': [], 'Day': [], 'Temperature avarage, Fahrenheit / Celsius': [], 'Temperature maximum, Fahrenheit / Celsius': [], 'Temperature minimum, Fahrenheit / Celsius': [],
           'Precipitation in / cm': [], 'Average daily wind speed (miles/hour)': [], 'Visibility in miles': [], 'Dew point, Fahrenheit / Celsius': [], 'Maximum sustained wind speed (miles/hour)': [],
           'Maximum wind gust reported for the day': [], 'Sea level pressure in mBARs': [], 'Station pressure in mBARs': [], 'Snow depth in inches': [], 'Weather type': []}
````

**Table parameters inside the website table**

````python
def update_dict(records, key_name, update_element):
  # rec = records.get(key_name)
  records[key_name].append(str(update_element))
````

````python
# Iterate through months and retrieve weather data for a specific location
for month in months:
    # Construct the URL for weather data retrieval
    url = (
        "https://geographic.org/global_weather/weather_data_2.php?month={0}"
        "&year={1}&id=171300-99999&path=weather_stations/129255_171869/"
        "171300-99999.txt&name={2}&country=Turkey"
    ).format(month, year, city)

    # Make an HTTP request to the URL
    page = requests.get(url)
    text = page.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(text, 'html.parser')

    # Find all table rows in the HTML content
    tr_tags = soup.find_all('tr')
    del tr_tags[0]  # Remove the header row

    check = 0
    # Iterate through table rows to extract weather information
    for table_row in tr_tags:
        # Iterate through desired parameters
        for param in parameters:
            # Find elements with the given parameter title
            row = table_row.findAll(attrs={'title': param})
            # Extract and update information in the records dictionary
            for info in row:
                update_dict(records, param, info.text)
                check = 1

        # Extract day number from the row
        day_number = table_row.find('strong').getText()

        # If parameter information was found, update additional details
        if check:
            update_dict(records, 'Day', day_number)
            update_dict(records, 'City', city)
            update_dict(records, 'Year', year)
            update_dict(records, 'Month', month)

        check = 0  # Reset the check flag
````

## 4.4  Making Query to the GitHub API:<a class="anchor" id="two-four"></a>

Let's send a query to retrieve and print some basic information about Linus Torvalds, fetched from the GitHub API.

````python
import requests

# The base URL for the GitHub API
BASE_URL = "https://api.github.com"

# Endpoint to get user information
USER_ENDPOINT = "/users/{}"

# The username you want to query information about
username = "torvalds"

# Sending GET request to the GitHub API
response = requests.get(BASE_URL + USER_ENDPOINT.format(username))

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    user_data = response.json()

    # Print some information about the user
    print(f"Name: {user_data['name']}")
    print(f"Bio: {user_data['bio']}")
    print(f"Location: {user_data['location']}")
    print(f"Public Repositories: {user_data['public_repos']}")
else:
    print(f"Failed to fetch data for {username}. Status code: {response.status_code}")

````

# 5. Ethical Considerations in Data Collection



In this final section, we emphasize the critical importance of ethical considerations in data collection, covering both legal compliance and broader ethical principles.

The data you deal with can be quite sensitive and understanding and adhering to laws like General Data Protection Regulation (GDPR), implementing robust data security measures, obtaining proper permissions, and practicing responsible data handling, you will be equipped to collect and use data in a way that respects individual privacy and societal norms.

## 5.1 Data Privacy and Security


### 5.1.1 Understanding GDPR and Other Privacy Laws

[General Data Protection Regulation](https://gdpr-info.eu/) (GDPR) is European law that governs the use and protection of personal data.
* It's key principles include **consent, transparency, data minimization, and the right to be forgotten.**

* Non-compliance can result in significant fines and legal consequences.

**Other Privacy Laws:**
Various countries have enacted laws similar to GDPR, such as the California Consumer Privacy Act (CCPA) in the U.S.
Knowing and complying with relevant laws in the jurisdictions where data is collected is essential.

**Ensuring Security in Data Collection**

**Data Encryption:** Protecting data during transmission and storage using encryption techniques.

**Access Controls:** Implementing proper controls to ensure that only authorized individuals can access sensitive data.

**Regular Audits and Monitoring:** Continuously monitoring data access and performing regular audits can help detect and prevent unauthorized access.

**Incident Response Planning:** Developing a plan to respond to data breaches to minimize harm and comply with legal obligations.

## 5.2 Ethical Conduct in Data Collection


### 5.2.1 Gaining Proper Permissions
* **Informed Consent:** Ensuring that participants understand the purpose of the data collection and voluntarily agree to participate.

* **Permission from Website Owners:** In the context of web scraping, adhering to [robots.txt](https://developers.google.com/search/docs/crawling-indexing/robots/intro#:~:text=A%20robots.txt%20file%20tells,or%20password%2Dprotect%20the%20page.) files and terms of service.

* **Compliance with Institutional Guidelines:** Following guidelines or obtaining approval from relevant ethical review boards, especially in research settings.

### 5.2.2 Responsible Data Handling
* **Data Minimization:** Collecting only the data that is necessary for the intended purpose.
* **Anonymization and Pseudonymization:** Using techniques to de-identify data, protecting individuals' privacy.
* **Data Retention Policies:** Defining and adhering to policies that govern how long data is retained and when it should be deleted.
* **Transparency with Stakeholders:** Clearly communicating with participants and stakeholders how data will be used, stored, and shared.
