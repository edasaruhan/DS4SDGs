## **2.1 Supervised Learning: Classification**

Classification in machine learning refers to a supervised learning approach where the goal is to predict the categorical class labels of new instances, based on past observations. These class labels are discrete, meaning they belong to a specific set and are not continuous.



### **2.1.1 Illustration of Classification**



Let's look at a simple classification problem example.

In the figure below, we observe a two-dimensional dataset. This means that each data point has two features. Each of these points is assigned to one of two classes, depicted by the colors blue and brown. Our goal in machine learning classification is to create a model that uses these features to predict the class (color) of a new, previously unseen data point.


![c1](https://jakevdp.github.io/PythonDataScienceHandbook/figures/05.01-classification-1.png)

We train our model to find the best position for this boundary, minimizing misclassifications. This process of determining the optimal boundary using the provided data is known as **learning the model**.

Please refer to the figure to see a visual representation of a trained classification model.

In essence, when we introduce a new data point without a label, the model will use its learned boundary to determine whether this point should be labeled as blue or red. This process of assigning labels to new data based on the model's understanding is referred to as **prediction**."

### **2.1.2 Classification Algorithms**

There are several algorithms for classification purpose. The most popular ones include the following:

**Logistic Regression:** Logistic regression is a machine learning algorithm for classification. In this algorithm, the probabilities describing the possible outcomes of a single trial are modelled using a logistic function.

**Naïve Bayes:** Naive Bayes algorithm based on Bayes’ theorem with the assumption of independence between every pair of features. Naive Bayes classifiers work well in many real-world situations such as document classification and spam filtering.

**K-Nearest Neighbours:** Neighbours based classification is a type of lazy learning as it does not attempt to construct a general internal model, but simply stores instances of the training data. Classification is computed from a simple majority vote of the k nearest neighbours of each point.

**Decision Tree:** Given a data of attributes together with its classes, a decision tree produces a sequence of rules that can be used to classify the data.

**Random Forest:** Random forest classifier is a meta-estimator that fits a number of decision trees on various sub-samples of datasets and uses average to improve the predictive accuracy of the model and controls over-fitting. The sub-sample size is always the same as the original input sample size but the samples are drawn with replacement.

**Support Vector Machine:** Support vector machine is a representation of the training data as points in space separated into categories by a clear gap that is as wide as possible. New examples are then mapped into that same space and predicted to belong to a category based on which side of the gap they fall.

* Look-ahead bias: Building a model based on data that is not supposed to be known.

It happens when a model is too simple to capture the underlying patterns in the data. To simplify, *poor performance on the training data and poor generalization to other data.*

**Evaluation of the Model**

One of the most fundamental tools used in machine learning classification tasks to summarize the performance of a model is the confusion matrix.
>**Confusion Matrix** presents a table that shows the predicted classes against the actual classes of a set of data points (as shown below). The confusion matrix helps you gain insights into the model's performance, especially when it comes to the types of errors it's making.


<div>
<img src="https://miro.medium.com/max/1400/1*N6I3pi0prhiJ_Y85HAx4wA.png" width="500"/>
</div>

Fundamentally, there are several standard metrics that we can used to evaluate the performance of model on our classification tasks: Accuracy, Precision, Recall, and F1-score.

> **Accuracy**\
Accuracy is one metric which gives the ratio of correctly classification data points among all data points. Let's assume that an example classification model predicts whether a patient has an underlying heart disease or not. Accuracy metric would give us the ratio between correctly classified patients and all patients.

<font size="4"><center>$$Accuracy = \frac{True Positive + True Negative}{Total}$$</center></font>

> **Precision & Recall**\
Precision is the ratio between the True Positives and all  Positives. For the example model above, Precision would be the measure of patients that we correctly identify having a heart disease out of all the patients actually having it.

<font size="4"><center>$$Precision = \frac{True Positive}{True Positive + False Positive}$$</center></font>
<br>

>**Recall** is the measure of our model correctly identifying True Positives. For our example, recall is the ratio between
correctly classified patients with a heart disease and all patients with a heart disease. Recall is sometimes referred as sensitivity.

<font size="4"><center>$$Recall = \frac{True Positive}{True Positive + False Negative}$$</center></font>

When to use precision or recall simply depends on the target of the classification model. If finding all positives is crucial, then the model should maximize recall so that it shouldn't misclassify any instance. If predicting positives correctly is crucial, then the model should maximize precision.

>**F1 Score**\
F1 score is a metric which combines precision and recall by taking the harmonic mean of the both. It is more useful than accuracy when the data is unevenly distributed among different labels.

<font size="4"><center>$$F1 Score = 2 * \frac{Precision * Recall}{Precision + Recall}$$</center></font>

#### **2.1.4.2 Introduce and Load Dataset**

* **Context**

* **Dataset Content**
>The *water_potability.csv* file (which is provided with this notebook) contains water quality metrics for 3276 different water bodies.

**1. pH value:**
PH is an important parameter in evaluating the acid–base balance of water. It is also the indicator of acidic or alkaline condition of water status. WHO has recommended maximum permissible limit of pH from 6.5 to 8.5. The current investigation ranges were 6.52–6.83 which are in the range of WHO standards.

**2. Hardness:**
Hardness is mainly caused by calcium and magnesium salts. These salts are dissolved from geologic deposits through which water travels. The length of time water is in contact with hardness producing material helps determine how much hardness there is in raw water. Hardness was originally defined as the capacity of water to precipitate soap caused by Calcium and Magnesium.

**3. Solids (Total dissolved solids - TDS):**
Water has the ability to dissolve a wide range of inorganic and some organic minerals or salts such as potassium, calcium, sodium, bicarbonates, chlorides, magnesium, sulfates etc. These minerals produced un-wanted taste and diluted color in appearance of water. This is the important parameter for the use of water. The water with high TDS value indicates that water is highly mineralized. Desirable limit for TDS is 500 mg/l and maximum limit is 1000 mg/l which prescribed for drinking purpose.

**4. Chloramines:**
Chlorine and chloramine are the major disinfectants used in public water systems. Chloramines are most commonly formed when ammonia is added to chlorine to treat drinking water. Chlorine levels up to 4 milligrams per liter (mg/L or 4 parts per million (ppm)) are considered safe in drinking water.

**5. Sulfate:**
Sulfates are naturally occurring substances that are found in minerals, soil, and rocks. They are present in ambient air, groundwater, plants, and food. The principal commercial use of sulfate is in the chemical industry. Sulfate concentration in seawater is about 2,700 milligrams per liter (mg/L). It ranges from 3 to 30 mg/L in most freshwater supplies, although much higher concentrations (1000 mg/L) are found in some geographic locations.

**6. Conductivity:**
Pure water is not a good conductor of electric current rather it is a good insulator. Increase in ions concentration enhances the electrical conductivity of water. Generally, the amount of dissolved solids in water determines the electrical conductivity. Electrical conductivity (EC) actually measures the ionic process of a solution that enables it to transmit current. According to WHO standards, EC value should not exceed 400 μS/cm.

**7. Organic_carbon:**
Total Organic Carbon (TOC) in source waters comes from decaying natural organic matter (NOM) as well as synthetic sources. TOC is a measure of the total amount of carbon in organic compounds in pure water. According to US EPA < 2 mg/L as TOC in treated / drinking water, and < 4 mg/Lit in source water which is use for treatment.

**8. Trihalomethanes:**
THMs are chemicals which may be found in water treated with chlorine. The concentration of THMs in drinking water varies according to the level of organic material in the water, the amount of chlorine required to treat the water, and the temperature of the water that is being treated. THM levels up to 80 ppm is considered safe in drinking water.

**9. Turbidity:**
The turbidity of water depends on the quantity of solid matter present in the suspended state. It is a measure of light emitting properties of water and the test is used to indicate the quality of waste discharge with respect to colloidal matter. The mean turbidity value obtained for Wondo Genet Campus (0.98 NTU) is lower than the WHO recommended value of 5.00 NTU.

**10. Potability:**
Indicates if water is safe for human consumption where 1 means Potable and 0 means Not potable.

#### **2.1.4.5 Separate Dataset into Features and Class** Model Training**

**Split the dataset into train and test sets**

````python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state=42)
````

**Define Classifiers that will be used on the dataset and initialize the classification algorithms**

````python
# Initialize a Logistic Regression model with a specified random state for reproducibility
lr = LogisticRegression(random_state=42)

# Initialize a K-Nearest Neighbors classifier
knn = KNeighborsClassifier()

# Initialize a Decision Tree classifier with a specified random state for reproducibility
dt = DecisionTreeClassifier(random_state=42)

# Initialize a Random Forest classifier with a specified random state for reproducibility
rf = RandomForestClassifier(random_state=42)

# Initialize an AdaBoost classifier with a specified random state for reproducibility
ada = AdaBoostClassifier(random_state=42)

# Initialize an XGBoost classifier with specified parameters
# - 'eval_metric': specifies the evaluation metric to be used
# - 'use_label_encoder': disables the use of label encoding to avoid associated warnings
# - 'random_state': ensures reproducibility of results
xgb = XGBClassifier(eval_metric = 'logloss', use_label_encoder=False, random_state=42)
````

````python
classifiers = [('Logistic Regression', lr), ('K Nearest Neighbours', knn),
               ('Decision Tree', dt), ('Random Forest', rf), ('AdaBoost', ada),
               ('XGBoost', xgb)]
````

**Train the classifiers with training set and test with the test set**

````python
# Initialize an empty list to store accuracy scores of each classifier
base_accuracy = []

# Loop over each classifier provided in the 'classifiers' list/tuple
# 'classifier_name' is the name or identifier of the classifier
# 'classifier' is the classifier object itself
for classifier_name, classifier in classifiers:

    # Fit the classifier on the training data
    classifier.fit(X_train, y_train)

    # Predict the outcomes on the test data
    y_pred = classifier.predict(X_test)

    # Calculate the accuracy of the predictions compared to the actual outcomes
    accuracy = accuracy_score(y_test, y_pred)

    # Print the classifier's name and its accuracy, formatted to 2 decimal places
    print('{:s} : {:.2f}'.format(classifier_name, accuracy))

    # Append the calculated accuracy to the 'base_accuracy' list
    base_accuracy.append(accuracy)
````

**Evaluation of the regression model:**

>**Mean Squared Error (MSE):**
The most common metric for regression tasks is MSE. It is the average of the squared difference between the predicted and actual value. Since it is differentiable and has a convex shape, it can be minimized easily.

<center><font size="4">$$ MSE = \frac{1}{n} \sum_{i=1}^{n}(y_i-\hat{y_i})^2$$</font></center>
MSE penalizes large errors for a single prediction because of the square term.

>**Mean Absolute Error (MAE):**
This is simply the average of the absolute value of the difference between the target value and the value predicted by the model. It is not preferred in cases where outliers are prominent.

<center><font size="4">$$ MAE = \frac{1}{n} \sum_{i=1}^{n}|y_i-\hat{y_i}|$$</font></center>
MAE does not penalize large errors as much as MSE.

> **Root Mean Squared Error (RMSE)**
This is the square root of the average of the squared difference of the predicted and actual value. R-squared error is better than RMSE. This is because R-squared is a relative measure while RMSE is an absolute measure of fit. Basically, RMSE is just the root of the average of squared residuals. We know that residuals are a measure of how distant the points are from the regression line. Thus, RMSE measures the scatter of these residuals.

<center><font size="4">$$ RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n}(y_i-\hat{y_i})^2}$$</font></center>
RMSE penalizes large errors because it is simply the square root of MSE.

>**R-squared**
This metric represents the part of the variance of the dependent variable explained by the independent variables of the model. It measures the strength of the relationship between your model and the dependent variable.

<center><font size="4">$$ R^2 = 1 - \frac{SS_{RES}}{SS_{TOT}} = 1 - \frac{\sum_{i=1}^{n}(y_i-\hat{y_i})^2}{\sum_{i=1}^{n}(y_i-\bar{y})^2}$$</font></center>

- If R² is high (say 1), then the model represents the variance of the dependent variable flawlessly.
- If R² is very low, then the model does not represent the variance of the dependent variable and regression is no better than taking the mean value, i.e. you are not using any information from the other variables.

* A Negative R² means you are doing worse than the mean value. It can have a negative value if the predictors do not explain the dependent variables at all such that RSS ~ TSS.
* Thus R² evaluates the scattered data points about the regression line.
It is not possible to see a model with an R² of 1. In that case, all predicted values are the same as actual values and this essentially means that all values fall on the regression line.

#### ⚠️ Note: Standardization

Before you fitting your training data to your regression model, it is good practive to always ensure that the independent variables are on a similar scale. This is because, when variables are on different scales, the regression coefficients are also on different scales, making them incomparable.

For instance, a coefficient for age (usually in years) and income (which could be in thousands or more) cannot be directly compared if the features aren't standardized. With standardization, all coefficients are on the same scale, and one can discern which features have more influence on the response variable.

**Standardization**, often referred to as *feature scaling* or *normalization*, is a preprocessing step that involves transforming the variables in a dataset to have a mean of zero and a standard deviation of one. This is achieved by subtracting the mean of each variable from each data point and then dividing by the standard deviation. Mathematically, if `x` is an original value, the standardized value `z` is given by:

$ z = \frac{x-mean}{standard \ deviation}$

For more information on fiting the Standard Scaler on Training Data, please visit the [sklearn website here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).

 ### ⚠️ Note:  

 Never fit the standard scaler to the test data. This is considered data leakage because you're allowing the model to "see" information from the test set during the training process. Always fit on the training data and then apply the transformation to any subsequent data (test, validation, or real-world data).

### **3.1.1 Basics of clustering**

One common case of unsupervised learning is "clustering" in which data is automatically assigned to some number of discrete groups. For example, we might have some two-dimensional data like that shown in the following figure:

![classification](https://jakevdp.github.io/PythonDataScienceHandbook/figures/05.01-clustering-1.png)

At first glance, it is clear that each of these points is part of a distinct group. Given this input, a clustering model will use the intrinsic structure of the data to determine which points are related. Using fast and intuitive **k-means algorithm**, we find the clusters shown in the following figure:

![classification](https://jakevdp.github.io/PythonDataScienceHandbook/figures/05.01-clustering-2.png)

k-means fits a model consisting of k cluster centers; the optimal centers are assumed to be those that minimize the distance of each point from its assigned center. Again, this might seem like a trivial exercise in two dimensions, but as our data becomes larger and more complex, such clustering algorithms need to be employed to extract useful information from the dataset.




### **3.1.2 Algorithms**

| Clustering Method             | Description                                                                                                           | Algorithms                     |
|-------------------------------|-----------------------------------------------------------------------------------------------------------------------|--------------------------------|
| Hierarchical Clustering       | Based on top-to-bottom hierarchy of the data points to create clusters                                                | DIANA, AGNES etc.              |
| Partitioning Methods          | Based on centroids and data points are assigned into a cluster based on its proximity to the cluster centroid         | k-means, k-medians, k-modes    |
| Distribution-based Clustering | Based on the probability distribution of the data, clusters are derived from various metrics like mean, variance etc. | Gaussian mixed models, DBCLASD |
| Density-based Clustering      | Based on density of the data points, also known as model based clustering                                             | DENCAST, DBSCAN                |
| Fuzzy Clustering              | Based on partitioning approach but data points can belong to more than one cluster                                    | Fuzzy C Means                  |
| Constraint-based Clustering   | Clustering is directed and controlled by user constraints                                                             | Decision trees, Random Forest  |

#### **Determining "K" number for clustering**

There are 2 main approaches for determining K number:

1. **Elbow Method:**

This is probably the most well-known method for determining the optimal number of clusters.

The elbow method involves calculating the Within-Cluster-Sum of Squared Errors (WSS) for various values of K and identifying the point at which the rate of reduction in WSS starts to diminish. This point is visually represented as an "elbow" in the plot of WSS against different values of K. The K value corresponding to this elbow point is often considered a reasonable choice for the number of clusters.

Within-Cluster-Sum of Squared Errors sounds a bit complex. Let’s break it down:

* The Squared Error for each point is the square of the distance of the point from its representation i.e. its predicted cluster center.
* The WSS score is the sum of these Squared Errors for all the points.
* Any distance metric like the Euclidean Distance or the Manhattan Distance can be used.

A sample plot for elbow:

![elbow](https://miro.medium.com/max/509/1*O_JmBi6rM6PlrPFx_uNrVQ.png)

To identify K value from plot after using elbow method, it is needed to look at the breakdown point. In this plot, it looks like an arm with a clear elbow at K = 3.

However, sometimes elbow method doesn't give clear plots like this.

![elbow2](https://miro.medium.com/max/875/1*t7VgOuC_Yp-Y1ND8z-eFtQ.png)

As you can see, the plot looks like a bit ambiguous compared with the first plot. Is the K 2 or 3? If we are not sure, we can also use The Silhouette Method.

2. **The Silhouette Method**

The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation).

The range of the Silhouette value is between -1 and +1. A high value is desirable and indicates that the point is placed in the correct cluster. If many points have a negative Silhouette value, it may indicate that we have created too many or too few clusters.

The Silhouette Value s(i) for each data point i is defined as follows:

$$ s(i) =
\begin{cases}
    {\large\frac{b(i)-a(i)}{max\{a(i),b(i)\}}} & , |C_i| > 1 \\
       0 & , |C_i| = 1
\end{cases} $$





#### ⚠️ Note:
> s(i) is defined to be equal to zero if i is the only point in the cluster. This is to prevent the number of clusters from increasing significantly with many single-point clusters.



Here, a(i) is the measure of similarity of the point i to its own cluster. It is measured as the average distance of i from other points in the cluster.
<br>
<center>For each data point $i \in C_i$  (data point i in the cluster $C_i$), let</center><br>
$$ a(i) = \frac{1}{|C_i|-1} \sum_{j\in C_i, i \neq j}^{} d(i,j) $$
<br>
<center>where d(i, j) is the distance between points i and j.</center>
<br>
Generally, Euclidean distance is used as the distance metric.
<br>

The Silhouette Score reaches its global maximum at optimal K. This should ideally appear as a peak in the Silhouette Value-versus-k plot.
Here is a sample plot:

![sil3](https://miro.medium.com/max/486/1*Otg9XRLPqOjb80Png6B9Og.png)

There is a clear peak at K = 3. Hence, it is optimal.

**3.1.3.5 Clustering (Finding Best K-value)**

**Elbow Method**

````python
# Initializing an empty list to store the distortions or sum of squared distances
distortions = []

# Define a range of values for 'k' (number of clusters) from 1 to 9
K = range(1, 10)

# Loop through each value of 'k'
for k in K:
    # Create a KMeans clustering model with the current 'k' value
    kmeanModel = KMeans(n_clusters=k, n_init=10)


    # Fit the model with the data (assumed to be stored in 'df')
    kmeanModel.fit(df)

    # Append the inertia (sum of squared distances from points to their cluster center) of the model to the distortions list
    distortions.append(kmeanModel.inertia_)
````

````python
# Set the figure size
plt.figure(figsize=(16,8))

# Plot the distortions for each value of K
plt.plot(K, distortions, 'bx-')

# Label x-axis as 'k'
plt.xlabel('k')

# Label y-axis as 'Distortion'
plt.ylabel('Distortion')

# Title of the plot
plt.title('The Elbow Method showing the optimal k')

# Display the plot
plt.show()
````

**Silhouette Method**

````python
# Import silhouette_score for evaluating the quality of clusters created using clustering algorithms
from sklearn.metrics import silhouette_score

# Loop over a range of cluster numbers from 2 to 9
for n_clusters in range(2, 10):

    # Initialize the KMeans clusterer with the current number of clusters
    clusterer = KMeans(n_clusters=n_clusters , n_init=10)


    # Fit the model to the data and predict the cluster labels for each data point
    preds = clusterer.fit_predict(df)

    # Get the cluster centers
    centers = clusterer.cluster_centers_

    # Compute the silhouette score for the current number of clusters
    score = silhouette_score(df, preds)

    # Print the silhouette score for the current number of clusters
    print("For n_clusters = {}, silhouette score is {})".format(n_clusters, score))
````

**Comment**: We are going to choose k-value as 2 since it has the highest silhouette score.

**Run Clustering with Best K-value**
