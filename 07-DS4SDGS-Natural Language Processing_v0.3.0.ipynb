### **2.1.1 Removing unnecessary characters**

The overarching goal of removing unnecessary characters from text is to strip out non-alphanumeric characters (like symbols) and often numbers, unless they're specifically relevant. This reduces noise, ensuring that the subsequent analyses and models are based on meaningful and relevant data, thereby improving their accuracy and effectiveness.

Let's define a function to clean dataset from unnecessary characters.

We will remove **digits** and **punctuations** from the text and only keep **letters.** Then we will be lowering the word cases due to provide a normalization.

**Why is text case lowering important in NLP tasks?**

There are several scenarios for which we might need to lower word cases. But, here's a simple example to demonstrate why it is relevant.

>Let's say we have a text with two sentences:
>1. It's my **life**
>2. **Life** is beautiful

The word "life" in the first sentence starts with lowercase "l" and in the second sentence it starts with the uppercase "L".

As the machines can't understand that they are the same words only by looking at them, we need to find a way to tell this to the machine. Ideally, the machine can't see the same word when it looks at **Life** and **life**. So basically, lowering this cases would be useful for avoiding an unnecessary differentiate.

````python
# **Define Text Cleaning Function**
# Importing the regular expressions library
import re

def clean_text(text_column):
    """
    Cleans the given column of text data by performing the following:
    1. Converts the text to lowercase.
    2. Removes any character that is not an alphabetical letter or a space.

    Parameters:
    - text_column (list or pandas Series): A column of text data.

    Returns:
    - list: A cleaned list of text data.
    """

    # Initialize an empty list to store cleaned text data
    texts = []

    # Loop through each text item in the provided column
    for text in text_column:
        # Convert the text to lowercase
        text = text.lower()

        # Use regular expressions to replace any character that isn't a-z, A-Z, or a space with an empty string
        text = re.sub('[^a-zA-Z ]+', '', text)

        # Append the cleaned text to the 'texts' list
        texts.append(text)

    # Return the cleaned texts list
    return texts

````

````python
osdg_dataset['cleaned_text'] = clean_text(osdg_dataset['text'])

processed_osdg_dataset = osdg_dataset[['text', 'cleaned_text']]
processed_osdg_dataset.head()
````

## **2.3 Tokenization**


Tokenization is a way of separating a piece of text into smaller units called tokens. Here, tokens can be either words, characters, or sub words.

 For example, consider the sentence:

`"Never give up"`

The most common way of forming tokens is based on `space`. Assuming space as a `delimiter`, the tokenization of the sentence results in `3 tokens`

 `Never - give - up `

As each token is a word, it becomes an example of **Word tokenization**.



Tokenization is a foundational pre-processing step . It involves seperating a piece of text into smaller units, commonly known as tokens. These tokens serve as the building blocks for more advanced processing and analysis.


Tokenization is a fundamental step in both traditional NLP methods like [Count Vectorizer](https://towardsdatascience.com/basics-of-countvectorizer-e26677900f9c) and Advanced Deep Learning-based architectures like [Transformers](https://en.wikipedia.org/wiki/Transformer_(machine_learning_model)).


Tokens can be words, characters, or sub-words.  

**Types of Tokens:**

* **Words:** In the case of word tokenization, the text is broken down into individual words. This is the most common form of tokenization and often involves using spaces as delimiters. For instance, the sentence `"Never give up"` becomes three separate tokens: `'Never', 'give', and 'up'`.

* **Characters:** When tokenizing by characters, every individual letter, punctuation mark, and numerical digit becomes its own token. The word `"Never"` would be represented as `'N', 'e', 'v', 'e', 'r'`.

* **Sub-words or Morphemes:** This is a more advanced form of tokenization where words are broken down into meaningful subunits. For instance, `"unhappiness"` might be broken into `'un-', 'happy', and '-ness'`. This approach is particularly useful for languages with compound words or for handling out-of-vocabulary words in machine learning models.

## **2.4 Stopword removal**

In Natural Language Processing (NLP), stop words refer to the most common words in a language that are generally filtered out before or during text processing. Examples in English include "and", "the", "is", "in", among others.


 While essential for human communication, these words might not contain significant meaning on their own, especially when analyzing text data.

 Removing stopwords helps to reduce noise, optimize computational resources, and improve the performance of text processing algorithms. However, their removal should be considered in the context of the specific NLP task at hand.

````python
# Importing the stopwords corpus from the nltk.corpus module
from nltk.corpus import stopwords

# Downloading the stopwords dataset using nltk's download function
# This is necessary if the stopwords dataset hasn't been previously downloaded.
nltk.download('stopwords')

# Loading English stopwords into the stop_words variable.
# stopwords.words('english') returns a list of common English words
# that are often removed during text preprocessing.
stop_words = stopwords.words('english')

# Displaying the first 20 stopwords to inspect and verify the loaded words.
stop_words[:20]
````

### **‚ö†Ô∏è Note: Extension of Stopwords**

Depending on the application, the list of stopwords can be extended beyond the default stopwords provided by the NLTK library.

Let's extend our stopword list to include additional stopwords by using the `append` or  the `extend` method on the list.



````python
# Additional stop words. Add as many as you want.
additional_stop_words = ["et al", "also", "likely", "still", "indeed", "would", "however" ]

# Extend the original list
stop_words.extend(additional_stop_words)

print(stop_words[-5:])  # print the last 5 words to check
````

````python
# Remove stopwords
a = processed_osdg_dataset['tokens'].to_list()

# Filter out the stopwords from the  list of tokens (a[0])
b = [word for word in a[0] if not word in stop_words] # a[0] is a list of our word tokens
````

````python
# Initialize an empty list to store tokenized texts after removing stopwords
tokens_without_stopwords = []

# Loop through each list of tokens from the 'tokens' column in processed_osdg_dataset
for tokens in processed_osdg_dataset['tokens'].to_list():

    # Use a list comprehension to filter out the stopwords from the current list of tokens
    # Append the result (a list of tokens without stopwords) to the 'tokens_without_stopwords' list
    tokens_without_stopwords.append([word for word in tokens if not word in stop_words])

# Add a new column 'tokens_without_stopwords' to the dataframe
# This column contains the lists of tokens after stopwords have been removed
processed_osdg_dataset['tokens_without_stopwords'] = tokens_without_stopwords

# Display the first three rows of the dataframe to inspect and verify the newly added column
processed_osdg_dataset.head(3)
````

## **2.4 Stemming**

Often, when searching for a particular word, it's beneficial to retrieve documents that contain its variations or related terms. Stemming is one of those methods that allow us to achieve this.

**Stemming** is a technique in Natural Language Processing (NLP) designed to reduce inflected or derived words to their word stem or root form.

 Essentially, it involves trimming the word to a base form by removing affixes (like suffixes) so that words that are derived from the same root can be identified as a common form.

This ensures that variations of a word are treated as the same, enhancing text consistency and search efficiency.

**Example:**

>am, are, is $\Rightarrow$ be

>car, cars, car's, cars' $\Rightarrow$ car

After applying stemming to a sentence, it could transform as follows:

> the boy's cars are different colors $\Rightarrow$ the boy car be differ color

Stemming typically employs heuristic methods, often by truncating words to capture the base form. This might involve removing endings or beginnings of words and recognizing common prefixes and suffixes in inflected words. The aim is to achieve a consistent base form for words, even though the stemmed word might not always be a valid or complete word in the language.


By reducing words to their base or root form, stemming simplifies text, facilitates more consistent text representation and enhances the performance of various text processing tasks.

Let's apply the process of stemming to an example dataset below.


````python
# Import the PorterStemmer class from the nltk.stem module
from nltk.stem import PorterStemmer

# Create an instance of the PorterStemmer class
ps = PorterStemmer()
````

````python
# Initialize an empty list to store the stemmed tokens for each sentence
stemmed_tokens = []

# Iterate through the 'tokens_without_stopwords' column of the 'processed_osdg_dataset' DataFrame
for tokens in processed_osdg_dataset['tokens_without_stopwords'].to_list():

    # Initialize an empty list to store the stemmed tokens for the current sentence
    stemmed_tokens_sentence = []

    # Iterate through each token in the current sentence
    for token in tokens:

        # Stem the token using the PorterStemmer instance 'ps' and append to the current sentence's list
        stemmed_tokens_sentence.append(ps.stem(token))

    # Append the stemmed tokens of the current sentence to the main list
    stemmed_tokens.append(stemmed_tokens_sentence)

# Add a new column named 'stems' to the 'processed_osdg_dataset' DataFrame
# This column contains the stemmed version of tokens for each sentence
processed_osdg_dataset['stems'] = stemmed_tokens

# Display the first 3 rows of the 'processed_osdg_dataset' DataFrame for verification
processed_osdg_dataset.head(3)
````

## **2.5 Lemmatization**


While stemming may seem enough, this indiscriminate cutting often causes variety of problems. Lemmatization comes as a more sophisticated approach to dealing with words.

**Lemmatization** does things properly with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the *lemma*. This lemma is usually the basic form of the regular dictionary instead of its stem, as lemma provides more information of the word itself.

**Example:**

studies $\Rightarrow$ third person, singular, present tense of the verb 'study' $\Rightarrow$ study

studying $\Rightarrow$ Gerund of the verb 'study' $\Rightarrow$ study

Lemmatization transforms words into their dictionary or meaningful base form which provides a more accurate and semantically rich representation of text, facilitates better information retrieval, improve text analysis, making it essential for various NLP tasks and applications.



````python
# Import the WordNetLemmatizer class from the nltk.stem module
from nltk.stem import WordNetLemmatizer

# Create an instance of the WordNetLemmatizer class
lemmatizer = WordNetLemmatizer()
````

````python
# Download the WordNet dataset. It's a large lexical database of English words.
# WordNetLemmatizer will use this dataset to find the base or dictionary form of words.
nltk.download('wordnet')

# Download the Open Multilingual WordNet (version 1.4) dataset.
# It's used to extend WordNet to languages other than English.
nltk.download('omw-1.4')

# Initialize an empty list to store the lemmatized tokens for each sentence
lemmas = []

# Iterate through the 'tokens_without_stopwords' column of the 'processed_osdg_dataset' DataFrame
for tokens in processed_osdg_dataset['tokens_without_stopwords'].to_list():

    # Initialize an empty list to store the lemmatized tokens for the current sentence
    lemmatized_tokens_sentence = []

    # Iterate through each token in the current sentence
    for token in tokens:

        # Lemmatize the token using the WordNetLemmatizer instance 'lemmatizer' and append to the current sentence's list
        lemmatized_tokens_sentence.append(lemmatizer.lemmatize(token))

    # Append the lemmatized tokens of the current sentence to the main list
    lemmas.append(lemmatized_tokens_sentence)

# Add a new column named 'lemmas' to the 'processed_osdg_dataset' DataFrame
# This column contains the lemmatized version of tokens for each sentence
processed_osdg_dataset['lemmas'] = lemmas

# Display the first 3 rows of the 'processed_osdg_dataset' DataFrame for verification
processed_osdg_dataset.head(3)
````

## **2.7 Parsing**

Parsing involves the process of determining the syntactic structure of a text. It involves breaking down a sentence into its individual components and establishing relationships between these components. The main goal is to convert a linear sequence of words from a text into a structure that represents its syntactic meaning.

Here are the primary types of parsing in NLP:

 **Constituency Parsing (or Phrase Structure Parsing)**

* This type of parsing breaks a text into sub-phrases, also known as constituents. These constituents can represent individual words or a group of words that act together as a single unit within a sentence.
* The output of a constituency parser is a tree with phrase labels like NP (Noun Phrase) and VP (Verb Phrase).
* The famous Penn Treebank dataset has annotated sentences using constituency parse trees, making it a popular resource for this kind of parsing.

**Dependency Parsing**

* This approach focuses on the grammatical relationships between words in a sentence, showing which words depend on (or modify) other words.
* The output is a tree (often a directed acyclic graph) where the nodes are words and the edges represent grammatical relationships with labels indicating the type of relationship (e.g., "subject", "object", "modifier").
* Dependency parsing is useful for tasks where understanding the precise relationship between words is crucial, like relation extraction or question answering.

**Semantic Parsing**

* This process maps a sentence to a formal meaning representation, often logic-based. This representation captures deeper, semantic relationships in the sentence beyond surface syntax.
* For instance, the sentence "John bought a car" might be represented in a logic form that identifies "John" as the buyer and "car" as the item bought.

**Morphological Parsing**

* This type of parsing breaks words down into their smallest meaningful units, called morphemes.
* For example, the word "unhappiness" can be broken down into three morphemes: "un-", "happy", and "-ness".

# **3 N-gram Language Models**

N-gram models are foundational in the domain of Natural Language Processing (NLP). They capture the sequence information of text, providing context and aiding in various linguistic tasks like text generation, spelling correction, and even machine translation.

An N-gram model breaks down text into chunks of 'n' consecutive words or characters. The primary types are:

* **Unigram**- Each word is considered individually.

>Example: [Let] [me] [explain] [to] [you.]

* **Bigram** - Pairs of consecutive words are taken.

> Example: [let me] [me explain] [explain to] [to you.]

* **Trigram**- Groups of three consecutive words.

> Example: [let me explain] [me explain to] [explain to you.]



## **3.1 What is N-gram?**

An n-gram is a contiguous sequence of n items from a given sample of text or speech. These 'items' can be words, characters, or even phonemes in speech recognition tasks.

![gram](https://devopedia.org/images/article/219/7356.1569499094.png)

**Applications of N-grams:**

* **Text Generation:** N-grams can help in predicting the next word in a sequence, facilitating text generation, especially in chatbots and auto-complete functionalities.

* **Spell Check and Correction:** By understanding the sequence and context of words, N-gram models can detect and correct spelling mistakes.

* **Machine Translation:** Translating text from one language to another can benefit from the contextual information provided by N-grams.

* **Text Classification:** In sentiment analysis or topic classification, the sequence of words can provide valuable insights, making N-grams essential.




#### **‚ö†Ô∏è Note:**
The efficiency of N-grams, however, is dependent on the choice of 'n'. A smaller 'n' might miss out on capturing broader context, while a very large 'n' might overfit to the specificities of a training dataset. Thus, finding an optimal 'n' is crucial for specific NLP tasks.

## **3.1 N-gram analysis**

Let us perform N-gram analysis by  extracting n-grams (bigrams, trigrams, etc.) from our text dataset.

````python
# Import the necessary function from NLTK to extract n-grams
from nltk.util import ngrams

# Define a function to extract n-grams from a given text
def get_ngrams(text, n):
    # Use the word_tokenize function to split the text into tokens, and then extract n-grams
    n_grams = ngrams(word_tokenize(text), n)
    # Join each n-gram tuple into a single string and return the list of these strings
    return [' '.join(grams) for grams in n_grams]

# Initialize empty lists to store bigrams, trigrams, and fourgrams for each text
bigrams = []
trigrams = []
fourgrams = []

# Iterate through the 'tokens_without_stopwords' column of the dataset

text_without_stopwords  = processed_osdg_dataset['tokens_without_stopwords'].apply(' '.join)

for text in text_without_stopwords:
    # For each text, extract bigrams, trigrams, and fourgrams and append them to their respective lists
    bigrams.append(get_ngrams(text, 2))
    trigrams.append(get_ngrams(text, 3))
    fourgrams.append(get_ngrams(text, 4))

# Add new columns to the dataset for bigrams, trigrams, and fourgrams
processed_osdg_dataset['bigrams'] = bigrams
processed_osdg_dataset['trigrams'] = trigrams
processed_osdg_dataset['fourgrams'] = fourgrams

# Display the first three rows of the updated dataset for verification
processed_osdg_dataset.head(3)
````

# **4 Part of Speech (POS)**

## **4.1 What is Part of Speech Tagging?**

Part-of-speech (POS) tagging is a fundamental process in Natural Language Processing (NLP) that assigns grammatical categories, or "tags", to individual words in a sentence. These tags denote the syntactic role a word plays in a sentence, such as whether it's a noun, verb, adjective, etc.

![sd](https://miro.medium.com/max/828/1*AyPfvdmnz5uF2UIiOX2ctA.png)

Part-of-speech tags describe the characteristic structure of lexical terms within a sentence or text, therefore, we can use them for making assumptions about semantics.

The process of POS tagging not only identifies the basic grammatical category of each word but also the more specific forms within those categories. For example, among verbs, the tags might further differentiate between past and present tense.

**Why is POS Tagging Important?**

* **Syntax Analysis:** POS tagging forms the foundation for parsing. This structure aids in understanding the relationships and hierarchies of different parts within the sentence.

* **Word Disambiguation:** Some words can serve as both nouns and verbs, or other combinations of parts of speech. For instance, "book" can be a noun ("a book") or a verb ("to book a flight"). POS tagging helps determine the correct interpretation based on context.

* **Improving Word Choice in Text Generation:** For applications that generate text, like chatbots or story generators, POS tagging can guide the system in choosing words that fit grammatically and semantically in the generated sentences.

* **Enhancing Search Engines:** POS tagging can enhance search results by allowing the engine to differentiate between different usages of a word, delivering more relevant results.

**Applications of POS Tagging**

They include:

* Named Entity Recognition
* Co-reference Resolution
* Speech Recognition
* Machine Translation

# **5 Named Entity Recognition(NER)**

## **5.1 What is Named Entity Recognition**

Named Entity Recognition (NER) is a sub-task in NLP that involves identifying and classifying named entities (such as names of persons, organizations, locations, dates, and other specific items) present in a given text into predefined categories.

The primary goal of NER is to extract structured information from unstructured textual data.

In practical context, named entities (specific, identifiable elements present in text) are often classified into the following predefined classes:
* **Person (PER)**: Names of individuals, e.g., "Alice," "Bob."
* **Organization (ORG)**: Names of institutions or companies, e.g., "OpenAI," "United Nations."
* **Location (LOC)**: Names of geographical locations like cities, countries, landmarks, e.g., "Paris," "Mount Everest."
* **Date (DATE)**: Dates or periods, e.g., "1998," "21st century."
* **Time (TIME)**: Times or durations, e.g., "2:00 PM," "5 minutes."
* **Monetary (MONEY)**: Monetary values, e.g., "$100," "5 euros."
* **Percentage (PERCENT)**: Percentages, e.g., "25%."




#### **‚ö†Ô∏è Note:**
These categories are general and therefore, depending on the specific domain and application, we can defined other custom categories.

## 5.1 **Applications of NER**

* **Information Extraction:** Extracting structured information from news articles, academic papers, or web pages for further processing or storing in databases.

* **Question Answering Systems:** To find precise answers to user queries, e.g., "Who is the CEO of OpenAI?"

* **Content Recommendation:** Recommending articles or news based on named entities mentioned in previously read articles by a user.

* **Sentiment Analysis:** Determining sentiments towards specific products, companies, or individuals.

* **Data Mining:** Quickly sifting through large volumes of data to find relevant information.

* **Document Summarization:** Identifying key entities in a text to produce concise summaries.
