### **1.2.1 Typical Neural Network Architecture:**

The typical neural network architecture consists of several layers.

#### **1.2.1 Input Layer**

The input layer is the initial layer through which the neural network receives its input X, which represents the data from which it learns.

The number of neurons (nodes) in this layer is typically equal to the number of features in the input data, X.

Ideally, each input neuron represents one element in the vector x.

Basically, the input layer  prepares the data for further processing by the subsequent layers.

#### **1.2.2 Output Layer**

This is last layer of neural network which outputs a vector y representing the result that the neural network came up with. The entries in this vector represent the values of the neurons in the output layer. Each neuron in the last layer would represent a different class.

 For instance, for a binary classification problem, you might have one output neuron, while for a multi-class classification problem, you might have as many neurons as there are classes.


#### **1.2.3 Hidden Layer**

In order to obtain a prediction vector y, the network must perform certain mathematical operations. These operations are performed in the layers between the input and output layers which called **Hidden Layers**.

These layers, aptly named because they are not directly observed from the input or output data, are pivotal in enabling deep learning models to represent and learn complex functions.

### üí° Fun Fact
The **"deep"** in deep learning or deep neural network usually refers to the multiple hidden layers found in the neural networks.

There are several types of hidden layers including: Dense (fully connected), convolutional layers, recurrent layer, normalization layer, etc.

* **1. Dense (Fully Connected) Layers:** These are the most common type where every neuron in the layer is connected to every neuron in the subsequent layer. The dense layer is a linear operation, but when combined with a non-linear activation function, it provides the network the capability to learn from the error and make adjustments, which is essential for learning complex patterns.

* **2. Convolutional Layers:** Primarily used in convolutional neural networks (CNNs), these layers are specifically designed to recognize spatial hierarchies in data, making them excellent for image recognition tasks. They work by sliding small windows (or kernels) over the input data to produce feature maps, capturing local patterns.

* **3. Recurrent Layers:** Found in recurrent neural networks (RNNs), these layers have loops that allow information to be passed from one step of the network to the next. This architecture is particularly beneficial for tasks where temporal dynamics and sequence recognition are crucial, like in natural language processing or time series prediction.

* **4. Normalization Layers:** These layers, such as Batch Normalization, adjust and scale the activations. By normalizing the layer's inputs, they can help maintain a stable mean and variance, improving the overall training speed and performance of deep networks by mitigating the problem of internal covariate shift.

## **1.2.2 Layer Connections:**

The figure below shows a simple neural network that consists only of the input and output layer with 2 input neurons and 3 output neurons:

![dl5](https://miro.medium.com/max/436/1*U6tiIuizR8PMkrOuqyH5vg.png)

Each connection between two neurons is represented by a numerical value, which called **weight(w)**. Each of these weight w has indices. The first value of the indices stands for the number of neurons in the layer from which the connection originates, the second value for the number of the neurons in the layer to which the connection leads.

All weights between two neural network layers can be represented by a matrix called the **weight matrix**.

![dl6](https://miro.medium.com/proxy/0*ebwG9bWKzhgZ0OLW)

A weight matrix has the same number of entries as there are connections between neurons. The dimensions of a weight matrix result from the sizes of the two layers that are connected by this weight matrix.

The number of rows corresponds to the number of neurons in the layer from which the connections originate and the number of columns corresponds to the number of neurons in the layer to which the connections lead.

In this particular example, the number of rows of the weight matrix corresponds to the size of the input layer which is two and the number of columns to the size of the output layer which is three.

### **1.2.3 Learning Process of a Neural Network:**

For a given input feature vector x, the neural network calculates a prediction vector, which we call here as h.

This step is also referred to as the forward propagation. With the input vector x and the weight matrix W connecting the two neural layers, we compute the dot product between the vector x and the matrix W.
The result of this dot product is again a vector, which is called z.

![dl6](https://miro.medium.com/proxy/0*LYoM4KMv_Y46Ko2c)

The final prediction vector h is obtained by applying a so-called activation function to the vector z. In this case, the activation function is represented by the letter Sigma.

### üí° **Activation Function**

An **activation function** is only a nonlinear function that performs a nonlinear mapping from z to h.

It allows the network to capture and represent complex relationships in the data.

Activation function is used to determine the output of neural network like yes or no. It maps the resulting values in between 0 to 1 or -1 to 1 etc. (depending upon the function)

There are a couple activation functions that are used in Deep Learning, which include:

1. tanh
2. sigmoid
3. Relu

At this point, the role of the weights unfolds well. A value of a neuron in a layer consists of a linear combination of neuron values of the previous layer weighted by some numeric values.

*These numerical values are the weights that tell us how strongly these neurons are connected with each other.*

During training, these weights are adjusted, some neurons become more connected, some neurons become less connected. As in a biological neural network, learning means the alteration of weights. Some weights make the predictions of a neural network closer to the actual ground truth vector y_hat, some weights increase the distance to the ground truth vector.

### **1.2.4 Loss Function:**

A loss function evaluates the difference between the neural networks predicted output (y) and the actual target values (y_hat).


Minimizing the loss function automatically causes the neural network model to make better predictions regardless of the exact characteristics of the task at hand.


You only have to select the right loss function for the task. Fortunately, there are only two loss functions that you should know (atleast for now) about to solve almost any problem that you encounter in practice.

Let's examine them below:


**1. Mean Squared Error (MSE) Loss**

Mean square error (MSE) also known as quadratic loss is typical used when solving regression problems. and it is defined as shown below.

![d11](https://miro.medium.com/max/875/0*hqIyWVt4QT9ocLf9)

The value of this loss function depends on the difference between y_hat and y.

A higher difference means a higher loss value, a smaller difference means a smaller loss value.

**Minimizing the loss function directly leads to more accurate** predictions of the neural network, as the difference between the prediction and the label decreases.



**2. Cross Entropy**

Cross entropy loss function is generally used when solving classification problems (binary and multi-class) classification problems.

Since the loss depends on the weights, we must find a certain set of weights for which the value of the loss function is as small as possible. The method of minimizing the loss function is achieved mathematically by a method called **gradient descent**

### **1.2.5 Gradient Descent:**

During gradient descent, we use the gradient of a loss function (or in other words the derivative of the loss function) to improve the weights of a neural network.

To understand the basic concept of the gradient descent process, let us consider a very basic example of a neural network consisting of only one input and one output neuron connected by a weight value w.

This means that the prediction is not accurate and we must use the gradient descent method to find a new weight value that causes the neural network to make the correct prediction. This value is the optimal weight parameter that would cause the neural network to make the correct prediction which is 6. In this case, the value for the optimal weight would be 3:

Our initial weight, on the other hand, is 5, which leads to a fairly high loss. The goal now is to repeatedly update the weight parameter until we reach the optimal value for that particular weight. This is the time when we need to use the gradient of the loss function. Fortunately, in this case, the loss function is a function of one single variable, which is the weight w:

This means that we have just used the gradient of the loss function to find out which weight parameters would result in an even higher loss value. But what we want to know is the exact opposite. Each time we are performing the update of the weights, we move down the negative gradient towards the optimal weights.

After each gradient descent step or weight update, the current weights of the network get closer and closer to the optimal weights until we eventually reach them and the neural network will be capable to do the predictions we want to make.


### **‚ö†Ô∏è Note:  Variants of the Gradient Descent Algorithm.**
By minimizing the loss, gradient helps in adjusting the wights of the neural network to improve its performance. The fundamental idea behind gradient descent is to adjust the weights in the direction of the steepest decrease in the lost function. This "direction" is given by the negative gradient of the loss with respect to the weights.

There are several variants of the Gradient Descent Algorithm. Let's examine the most common types:

**1. Batch Gradient Descent**

Batch Gradient Descent uses the entire dataset to compute the gradient of the loss function and update the model's weights in one step.

* **Pros:**
 * Stable convergence.
 * Benefits from vectorized operations.
* **Cons:**
 * Can be very slow for large datasets.
 * Memory-intensive with large datasets.

**2. Stochastic Gradient Descent (SGD):**

Updates the model's weights using only one training example at a time.
* **Pros:**
 * Faster weight updates.
Can escape local minima due to noisy updates.
* **Cons:**
 * Much noisier updates, leading to oscillations.
 * Typically requires more epochs to converge.

**3. Mini-Batch Gradient Descent:**
A compromise between Batch and SGD. Uses a mini-batch (subset) of training examples for each update.

* **Pros:**
 * Faster than batch gradient descent for large datasets.
 * Can benefit from hardware optimizations.
* **Cons:**
 * Still can have noisy updates (though less than pure SGD).
 * Mini-batch size requires tuning.

 ### **More Advanced variants of Gradient Algorithm**

**4. Momentum Gradient Descent**
 Incorporates a velocity term, which is a fraction of the previous update, to determine the next update.
* **Pros:**
 * Accelerates convergence.
 * Reduces oscillations.
* **Cons:**
 * Introduces an additional hyperparameter.

**5. Adagrad**
Adapts the learning rate for each parameter based on the historical gradient information.
* **Pros:**
 * Less sensitive to initial learning rate.
 * Can handle sparse data.
* **Cons:**
 * The learning rate can decay too aggressively and become very small.

**6. RMSprop**
Modifies Adagrad by computing a moving average of squared gradients to adjust the learning rate.
* **Pros:**
 * Overcomes Adagrad's aggressive decay.
* **Cons:**
 * Introduces another hyperparameter.

**7. Adam (Adaptive Moment Estimation)**
Combines Momentum and RMSprop. Keeps moving averages of both gradients and squared gradients.
* **Pros:**
 * Typically works well in practice with little tuning.
* **Cons:**
 * Might be sensitive to hyperparameters in some scenario

### **1.2.6 Curse of Dimensionality:**

Before we build a neural network, let's mention a critical concept known as Curse of Dimensionality.

Curse of Dimensionality describes the explosive nature of increasing data dimensions and its resulting exponential increase in computational efforts required for its processing and/or analysis. This term was coined by Richard Bellman in the 1960s.

 In the context of deep learning and neural networks, the curse of dimensionality can manifest in several ways:

 * **Data Sparsity:** In high-dimensional spaces, data can become sparse. This means that to get representative samples of all possible combinations of features, you need an exponentially increasing amount of data. For instance, if 10 samples provide a good representation of a single feature, you'd need 100 samples for two features, 1,000 for three features, and so on.

* **Distance Measures Lose Meaning:** In high dimensions, the distance between most pairs of points tends to be very similar, which can be problematic for clustering or nearest neighbor searches.

* **Overfitting:** High-dimensional datasets increase the risk of overfitting, especially if you have a limited amount of data. This is because in higher dimensions, there's a higher likelihood that your model will fit to the noise in the training data rather than the actual underlying pattern.

* **Computational Complexity:** Many algorithms that work fine in low dimensions become intractable as the dimensionality increases.

Hughes (1968) in his study concluded that with a fixed number of training samples, the predictive power of any classifier first increases as the number of dimensions increase, but after a certain value of number of dimensions, the performance deteriorates. Thus, the phenomenon of curse of dimensionality is also known as Hughes phenomenon.


image.png

[Representation of Hughes Principle](https://towardsdatascience.com/curse-of-dimensionality-a-curse-to-machine-learning-c122ee33bfeb)


**Methods to Handle the Curse of Dimensionality in Deep Neural Networks:**

 The methods to handle the curse of dimensionality include but not limited to the following:

* **Dimensionality Reduction**: Techniques like PCA (Principal Component Analysis), t-SNE, and autoencoders can be used to reduce the number of features in the data without losing much information.

* **Regularization**: Techniques such as L1 and L2 regularization can help prevent overfitting by adding penalty terms to the loss function, which in turn constrain the weights of the model. Dropout is another regularization technique often used in deep neural networks which randomly deactivates a fraction of neurons during training.

* **Feature Engineering**: Manually or algorithmically selecting the most important features or transforming features into a more useful representation can reduce dimensionality.

* **Feature Extraction**: Deep learning itself can be used as a form of feature extraction. For instance, convolutional neural networks (CNNs) are adept at this, transforming raw image data into a hierarchical set of features.

### **1.2.6 Building a Neural Network with Keras Library**

[Keras](https://keras.io/) is an open-source library that provides a
Python interface for building artificial neural networks.

Let's use Keras library to construct a simple feedforward neural network model, consisting of three layers.


````python
#remove any potential warnings
import warnings
warnings.filterwarnings('ignore')
````

````python
#Dependencies
import keras  # Importing the Keras library
from keras.models import Sequential  # Importing the Sequential model class from Keras
from keras.layers import Dense  # Importing the Dense layer class from Keras

# Neural network
model = Sequential()  # Initializing a sequential neural network model
model.add(Dense(16, input_dim=20, activation='relu'))  # Adding a dense layer with 16 neurons, input dimension of 20, and ReLU activation
model.add(Dense(12, activation='relu'))  # Adding another dense layer with 12 neurons and ReLU activation
model.add(Dense(4, activation='softmax'))  # Adding the output layer with 4 neurons and softmax activation
````

**Model Description**

* **Sequential:** specifies to keras that we are creating model sequentially and the output of each layer we add is input to the next layer we specify.

* **model.add:** is used to add a layer to our neural network

* **Dense:** is used to specify the fully connected layer.

The arguments of Dense are output dimension which is 16 in the first case, input dimension which is 20 for input dimension and the activation function to be used which is **relu** in this case.

The second layer is similar, we dont need to specify input dimension as we have defined the model to be sequential so keras will automatically consider input dimension to be same as the output of last layer i.e 16.

In the third layer(output layer) the output dimension is 4(number of classes).

Now as we have discussed earlier, the output layer takes different activation functions and for the case of multiclass classification, it is **softmax**.

Let us now compiled the model we just built.

````python
# Compiling the model by specifying the loss function, optimizer, and metrics to monitor
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
````

Here loss is categorical_crossentropy loss. Categorical_crossentropy specifies that we have multiple classes.

The optimizer is **Adam**. Metrics is used to specify the way we want to judge the performance of our neural network. Here we have specified it to accuracy.

After these steps you are ready train your models.

### üí° **Transfer Learning**

**What is Transfer Learning?**  
Transfer learning is a technique in machine learning where a model developed for a task is reused as the starting point for a model on a second task. This allows the second model to learn faster and achieve better performance because it can leverage the knowledge gained from the first task.

**Why is it used?**  
Transfer learning is used because it can save a lot of time and resources compared to training a model from scratch. Additionally, it can be used when there is limited data available for the second task, making it easier to train a good model.

**When should we use it?**  
Transfer learning should be used when the second task is related to the first task in some way. For example, if you have a model that has been trained to recognize objects in images and you want to train a model to recognize objects in videos, you can use the image recognition model as a starting point for the video recognition model.

**Transfer Learning approaches**  
There are several approaches to transfer learning, such as fine-tuning, feature extraction, and multi-task learning. Fine-tuning involves training a pre-trained model on the new task by adjusting the weights of the model. Feature extraction involves using the features learned by the pre-trained model and training a new classifier on top of it. Multi-task learning involves training a model on multiple tasks simultaneously, which can lead to better performance on both tasks.

````python
# Import necessary libraries
import re
import os
import pandas as pd

# Function to add class name prefix to the file path in the dataframe
def add_class_name_prefix(df, col_name):
    # Use regular expressions to find the first occurrence of a digit in the string,
    # then add a prefix based on everything before that digit, followed by the original string.
    df[col_name] = df[col_name].apply(lambda x: x[:re.search("\d", x).start()] + '/' + x)
    return df

# Function to map class ID to its corresponding label
def class_id_to_label(id):
    label_map = {1: 'glass', 2: 'paper', 3: 'cardboard', 4: 'plastic', 5: 'metal', 6: 'trash'}
    return label_map[id]

# Define directory and file paths


basedir = "/content/drive/MyDrive/SDG_AI_LAB_UNDP/DS4SDGS"

IMAGES_DIR = basedir + '/data/raw/garbage_classification/Garbage classification/'
train_file = basedir + '/data/raw/garbage_classification/one-indexed-files-notrash_train.txt'
val_file   = basedir + '/data/raw/garbage_classification/one-indexed-files-notrash_val.txt'
test_file  = basedir + '/data/raw/garbage_classification/one-indexed-files-notrash_test.txt'

# Read data files into Pandas DataFrames
df_train = pd.read_csv(train_file, sep=' ', header=None, names=['rel_path', 'label'])
df_valid = pd.read_csv(val_file,   sep=' ', header=None, names=['rel_path', 'label'])
df_test  = pd.read_csv(test_file,  sep=' ', header=None, names=['rel_path', 'label'])

# Apply the add_class_name_prefix function to modify 'rel_path' in each DataFrame
df_train = add_class_name_prefix(df_train, 'rel_path')
df_valid = add_class_name_prefix(df_valid, 'rel_path')
df_test  = add_class_name_prefix(df_test,  'rel_path')

# Convert the class IDs to their corresponding labels
df_train['label'] = df_train['label'].apply(class_id_to_label)
df_valid['label'] = df_valid['label'].apply(class_id_to_label)
df_test['label']  = df_test['label'].apply(class_id_to_label)

# Print the number of samples in each dataset
print(f'Found {len(df_train)} training, {len(df_valid)} validation and {len(df_test)} samples.')
````

#### **2.3.1 Extract images with ImageDataGenerator for Training**

````python
# Importing ImageDataGenerator from TensorFlow's Keras API
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an instance of the ImageDataGenerator class
datagen = ImageDataGenerator()

# Configure the data generator for the training dataset
datagen_train = datagen.flow_from_dataframe(
    dataframe=df_train,          # Dataframe containing the file paths and labels
    directory=IMAGES_DIR,        # Root directory where images are stored
    x_col='rel_path',            # Column in dataframe containing the file paths
    y_col='label',               # Column in dataframe containing the labels
    color_mode="rgb",            # Images will be converted to have 3 color channels (Red, Green, Blue)
    class_mode="categorical",    # Labels are in "categorical" format (useful for multi-class classification)
    batch_size=32,               # Number of samples to include in each batch (default is 32)
    shuffle=True,                # Shuffle the dataset randomly
    seed=7                       # Random seed for reproducibility
)

# Configure the data generator for the validation dataset
datagen_valid = datagen.flow_from_dataframe(
    dataframe=df_valid,          # Dataframe containing the file paths and labels
    directory=IMAGES_DIR,        # Root directory where images are stored
    x_col='rel_path',            # Column in dataframe containing the file paths
    y_col='label',               # Column in dataframe containing the labels
    color_mode="rgb",            # Images will be converted to have 3 color channels (Red, Green, Blue)
    class_mode="categorical",    # Labels are in "categorical" format (useful for multi-class classification)
    batch_size=32,               # Number of samples to include in each batch (default is 32)
    shuffle=True,                # Shuffle the dataset randomly
    seed=7                       # Random seed for reproducibility
)
````

## **2.4 Model Training**

**Model Explanation:**

* We will be using **ResNet50** network

ResNets make it possible to train ultra deep neural networks. It has a network that can contain hundreds or thousands of layers and still achieve great performance.

Our model architecture has the following properties:

* Activation function as **relu** in hidden layer
* Activation function as **softmax** as the output layer
* Optimizer as **adam**
* Loss function as **adam**
* Metric as **categorical accuracy**

````python
# Importing TensorFlow
import tensorflow as tf
from tensorflow.keras.applications import ResNet50

# Function to build the model
def build_model(num_classes):
    # Load the ResNet50 architecture with pretrained weights from ImageNet but without the top (final fully connected) layers
    base_model = ResNet50(weights='imagenet', include_top=False)

    # Get the output tensor of the base model
    x = base_model.output

    # Add a Global Average Pooling 2D layer to reduce the spatial dimensions of the output volume
    x = tf.keras.layers.GlobalAveragePooling2D()(x)

    # Add a fully-connected layer with 1024 hidden units and ReLU activation
    x = tf.keras.layers.Dense(1024, activation='relu')(x)

    # Add a final fully-connected layer with softmax activation for class probabilities
    predictions = tf.keras.layers.Dense(num_classes, activation='softmax')(x)

    # Configure the inputs and outputs and compile the model
    model = tf.keras.Model(inputs=base_model.input, outputs=predictions)

    # Freezing the layers of the base model (ResNet50)
    for layer in base_model.layers:
        layer.trainable = False

    return model

# Build the model with 6 output classes
net = build_model(num_classes=6)

# Compile the model with Adam optimizer, categorical cross-entropy loss and categorical accuracy as the metric
net.compile(optimizer='Adam',
            loss='categorical_crossentropy',
            metrics=[tf.keras.metrics.categorical_accuracy])

# Display the architecture of the model
net.summary()
````

````python
history = net.fit(datagen_train,
    #generator=datagen_train,
    validation_data=datagen_valid,
    epochs=20,
    validation_freq=1)
````

````python
# Import TensorFlow
import tensorflow as tf

# Create a ModelCheckpoint callback
# This callback saves the model at the filepath specified whenever 'val_accuracy' improves
# Reference: https://keras.io/api/callbacks/model_checkpoint/
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    basedir + "/data/raw/garbage_classification/resnet50_model.h5",  # Filepath where the model will be saved
    monitor="val_accuracy",                              # Metric to monitor during training
    save_best_only=True,                                 # Only save the model that has the best 'val_accuracy'
    mode="auto",                                         # Mode set to 'auto', which means it would figure out when to save the model based on the monitored metric (val_accuracy in this case)
    verbose=1                                            # Logs the checkpoints being saved during training
)
````

````python
# Callback that streams epoch results to a CSV file -- Reference: https://keras.io/api/callbacks/csv_logger/
log = tf.keras.callbacks.CSVLogger(basedir + '/data/raw/garbage_classification/resnet50_model.txt')
````

* We will be using **early stopping** to avoid overfitting
* Monitoring the **validation loss**
* **Epoch** number as 20(this parameter can be fine tuned after reviewing the outputs)

````python
# Import TensorFlow
import tensorflow as tf

# Create an EarlyStopping callback
# This stops the training if 'val_loss' does not improve after 5 epochs
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',            # Metric to monitor during training
    patience=5,                    # Number of epochs with no improvement to wait before stopping the training
    verbose=1,                     # Print messages for stopping
    restore_best_weights=True      # Restore model weights from the epoch with the best value of the monitored metric
)

# Training the model
history = net.fit(
    datagen_train,                 # Training data generator
    validation_data=datagen_valid, # Validation data generator
    epochs=20,                     # Number of epochs to train for
    validation_freq=1,             # Frequency (in epochs) at which to compute validation metrics
    callbacks=[early_stop, checkpoint, log] # List of callbacks to apply during training
)
````

````python
# Import Matplotlib library for plotting
import matplotlib.pyplot as plt

# Create a 1x2 grid of subplots. The figure size is set as (20, 4)
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 4))

# Plot the training loss on the first subplot
axs[0].plot(history.history['loss'], label='loss')

# Plot the validation loss on the first subplot
axs[0].plot(history.history['val_loss'], label='val_loss')

# Plot the training accuracy on the second subplot
axs[1].plot(history.history['categorical_accuracy'], label='acc')

# Plot the validation accuracy on the second subplot
axs[1].plot(history.history['val_categorical_accuracy'], label='val_acc')

# Add a legend to identify the lines in both subplots
plt.legend()

# Display the plots
plt.show()
````

# **3 Introduction to Transformers**


Transformers are a type of neural network architecture that were introduced in a 2017 paper by Google researchers, ["Attention is All You Need"](https://proceedings.neurips.cc/paper_files/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf).

They have since become popular in a wide range of natural language processing (NLP) tasks such as language translation, text summarization, and question answering.  It brought a significant shift in the way sequence-to-sequence tasks (like translation) were approached.

Prior to Transformers, recurrent neural networks (RNNs) and their variants, like long short-term memory (LSTM) and gated recurrent units (GRU), were commonly used for such tasks. The Transformer model, however, does away with recurrence entirely and relies solely on attention mechanisms to draw global dependencies between input and output.

To read more about the Transformer model architecture, click the link to the research paper ["Attention is All You Need"](https://proceedings.neurips.cc/paper_files/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf). Also consult  the Machine Tutor section below:

## **3.1. BERT Language Model**

Pre-training refers to using model parameters of tasks that have been learned before to initialize the model parameters of new tasks. In this way, the old knowledge helps new models successfully perform new tasks from old experience instead of from scratch.

BERT's pre-training task is called MLM, or Masked Language Model. In the input word sequence of this model, 15% of the words are randomly masked and the task is to predict what they are. What we see is that BERT can predict these words from both directions‚Ää-‚Äänot just left-to-right or right-to-left.

There are significant benefits to using a pre-trained model. It reduces computation costs, and your carbon footprint, and allows you to use state-of-the-art models without having to train one from scratch. Transformers provides access to thousands of pre-trained models for a wide range of tasks. When you use a pre-trained model, you train it on a dataset specific to your task.

This is known as fine-tuning, an incredibly powerful training technique. You can fine-tune a pretrained model with a deep learning framework such as TensorFlow with Keras, and PyTorch.

BERT was pre-trained using only an unlabeled, plain text corpus (namely the entirety of the English Wikipedia, and the Brown Corpus). It continues to learn unsupervised from the unlabeled text and improve even as its being used in practical applications (ie Google search). Its pre-training serves as a base layer of "knowledge" to build from. From there, BERT can adapt to the ever-growing body of searchable content and queries and be fine-tuned to a user's specifications. This process is known as transfer learning.
